```python
############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################

############################################################
# All imports
###########################################################
from typing import List
from time import process_time 
import collections

def xxxx(ladders, snakes):
    num_moves = [-1]
    work = [0]
    max = 100
    p = Exam(max,ladders, snakes,num_moves,work,False)
    return num_moves[0]

    
class Exam:
    def __init__(self, max:'int', ladders:'list', snakes:'list', num_moves:'List of size 1',work:'List of size 1',show:'boolean'):
        self._max = max #size of board
        self._ladders = ladders
        self._snakes = snakes
        self._num_moves = num_moves
        self._work = work
        self._show = show

        ## you can initialize all your private variables here STARTS
        #initialize path
        self._route = {i:[] for i in range(1,self._max+1)}
	    ## you can initialize all your private variables here ENDS
        self._solution()


    def _set_num_moves(self, m:'int'):
        self._num_moves[0] = m 

    def _increment_work(self):
        self._work[0] = self._work[0] + 1

    def _solution(self):
        #MUST WRITE 2 routines
        self._alg()
        if (self._show):
            # 2) Print shortest path
            #ONLY 50 marks will be given if you cannot print the shortest path
            print("The paths from 1 to",self._max, "are as follows")
            self._print_all_shortest_path()
            print("WORK DONE =",self._work[0])

    ############################################################
    # WRITE CODE BELOW
    ###########################################################
              
    def _alg(self):
        #Initialize the board with 0s
        game = [0] * (self._max + 1)

    #Update board with ladders and snakes
        for i, j in self._ladders + self._snakes:
            game[i] = j

    #Initialize distance with infinity except for starting square
        distance = [float("inf")] * (self._max + 1)
        distance[1] = 0

        # Create a priority queue for A* and add the starting square
        queue = [(0, 1)]

        while queue:
            # Pop the element with the lowest f-value from the queue
            f, square = heappop(queue)

        #Consider all squares reachable from the current square
            for new in range(square + 1, square + 7):
            # If the square is out of game, skip it
                if new > self._max:
                    continue

            #If the square has a ladder or a snake, update the next square accordingly
                if game[new] != 0:
                    new = game[new]

            #Calculate the g-value (distance from start) and h-value (heuristic estimate to goal) for the next square
                g = distance[square] + 1
                h = (self._max - new) // 6

                # If the f-value for the next square is greater than the current f-value plus 1
                if g + h < distance[new]:
                    # Update the distance to the next square
                    distance[new] = g
                    # Update the path to the next square
                    self._route[new] = self._route[square] + [square]
                    # Add the next square to the queue with its f-value as priority
                    heappush(queue, (g + h, new))
                    # Increment the work counter
                    self._increment_work()

    # Set the number of moves required to reach the last square
        self._set_num_moves(distance[self._max])

        
    def _print_all_shortest_path(self):
        # For all squares except the first square
        for x in range(2, self._max + 1):
         #If there exists a path to the target square
            if self._route[x]:
        #Print the path and its length
                path_str = "->".join(str(x) for x in self._route[x][1:] + [x])
                print(f"1->{path_str} Length = {len(self._route[x])}")
            else:
            # Print that there is no path to the target square
                print(f"1->{x} NO Path")
```
