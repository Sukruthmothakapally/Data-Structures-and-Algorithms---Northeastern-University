```python
############################################################
# BuildTree.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2022
###########################################################

class BuildTree():
    def __init__(self):
        pass

    ############################################################
    # build tree from Python list
    # Time: THETA(n)
    # Space: THETA(n)
    # WRITE CODE BELOW
    ###########################################################
    def build_tree_from_list(self,l:'input list')->'root':
        root = None
        n = len(l)
        if (n):
            list_of_nodes = []
            for e in l:
                if (e):
                    node = TreeNode(e)
                    list_of_nodes.append(node)
                else:
                    list_of_nodes.append(None)
      
            pos = 0 
            assert(list_of_nodes[0])
            root = list_of_nodes[0];
            for i in range(n):
                nodei = list_of_nodes[i]
                if (nodei):
                    j = pos * 2 + 1
                    #print("Building ",i, pos, "Left Kid",j,"Right Kid", j+1)
                    if (j < n):
                        nodej = list_of_nodes[j]
                        if (nodej):
                            nodei.left = nodej
                    j = j + 1
                    if (j < n):
                        nodej = list_of_nodes[j]
                        if (nodej):
                            nodei.right = nodej
                    pos = pos + 1 #Keep pos ready for next reading
        return root 
```
